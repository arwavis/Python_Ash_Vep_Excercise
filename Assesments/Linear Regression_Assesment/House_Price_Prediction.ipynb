# Data Preprocessing 
### Converting categorical variables of the dataset into numerical variables - using ONE HOT ENCODING technique

```python
data.head()
```

```python
data['mainroad']=data['mainroad'].apply({'yes':1,'no':0}.get)
data['guestroom']=data['guestroom'].apply({'yes':1,'no':0}.get)
data['basement']=data['basement'].apply({'yes':1,'no':0}.get)
data['hotwaterheating']=data['hotwaterheating'].apply({'yes':1,'no':0}.get)
data['airconditioning']=data['airconditioning'].apply({'yes':1,'no':0}.get)
data['prefarea']=data['prefarea'].apply({'yes':1,'no':0}.get)
data['furnishingstatus']=data['furnishingstatus'].apply({'furnished':1,'semi-furnished':0, 'unfurnished':2}.get)


data.head()
```

# Dividing the dataset into dependent and independent columns

```python

X=data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']] #independent features from the dataset

y=data['price'] #dependent column from the dataset
```

```python
X.head()
```

```python
y.head()
```

# Splitting the dataset into training and testing set
## 20% of the dataset will be used for testing(evaluation) and 80% of the data will be used for training purposes

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#creating the machine learning model
from sklearn.linear_model import LinearRegression
linearmodel=LinearRegression()
linearmodel.fit(X_train,y_train) # training is taking place
```

```python
#predictions
linearmodel.predict(X_test)
```

```python
#calculating the accuracy of the model
linearmodel.score(X,y) # evaluation is taking place
```

```python
X.head()
```

```python
#predictions with respect to new dataset
data_new={'area':9520,'bedrooms':3,'bathrooms':2,'stories':2,'mainroad':0,'guestroom':1,'basement':0,'hotwaterheating':1,'airconditioning':1,'parking':1,'prefarea':0,'furnishingstatus':1}
index=[1]#serial number
my_data=pd.DataFrame(data_new,index)
```

```python
print(my_data)
```

```python
#Pricing allocated
my_data_price=linearmodel.predict(my_data)
rounded_price = np.round(my_data_price, 2)
print(f"the charges allocated to the new customer is {rounded_price}")
```
